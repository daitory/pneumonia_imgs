```python
def directory_imgstodf(normal_imgs,path_normal,dimension):  

    pos=[i for i in np.arange(0,dimension,3)]
    start=datetime.datetime.now()



    normaldf=[]

    for i in range(0,len(normal_imgs)):
        print(i)
        print(normal_imgs[i])
        if len(re.findall("jpeg",normal_imgs[i]))>0:
            #imgs_normal=cv2.imread(path_normal + '/' +normal_imgs[i])
            #imgs_normal2=cv2.resize(imgs_normal,(256,256))
            normaldf.append(path_normal + '/' +normal_imgs[i])
            

    print("time elapsed", (datetime.datetime.now()-start) )
    return(normaldf)
```

```python
dimension1=256*256*3
normal_lst=directory_imgstodf(normal_imgs,path_normal,dimension1)
pneumonia_lst=directory_imgstodf(pneunmonia_imgs,path,dimension1)
```

```python
normal_train=pd.DataFrame(normal_lst)
pneumonia_train=pd.DataFrame(pneumonia_lst)
normal_train['Target']='0'
pneumonia_train['Target']='1'


normal_train_val=pd.DataFrame(normal_lst_val)
pneumonia_train_val=pd.DataFrame(pneumonia_lst_val)
normal_train_val['Target']='0'
pneumonia_train_val['Target']='1'



normal_train_test=pd.DataFrame(normal_lst_test)
pneumonia_train_test=pd.DataFrame(pneumonia_lst_test)
normal_train_test['Target']='0'
pneumonia_train_test['Target']='1'
```

```python
TrainDF=pd.concat([pneumonia_train.iloc[range(0,normal_train.shape[0]),:],normal_train],axis=0)
ValDF=pd.concat([pneumonia_train_val,normal_train_val],axis=0)
TestDF=pd.concat([pneumonia_train_test,normal_train_test],axis=0)
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255,  rotation_range=45, width_shift_range=.15,
                    height_shift_range=.15,
                    horizontal_flip=True,
                    zoom_range=0.75                 
                                          ) 
```

```python
TrainDF.columns=['filename','Target']
ValDF.columns=['filename','Target']
TestDF.columns=['filename','Target']
```

```python
TrainDF1=TrainDF.reset_index()
ValDF1=ValDF.reset_index()

TestDF1=TestDF.reset_index()
```

```python
batch_size = 16
epochs = 15
IMG_HEIGHT = 256
IMG_WIDTH = 256
```

```python
train_data_gen=train_image_generator.flow_from_dataframe(TrainDF1,x_col='filename',y_col='Target',target_size=(IMG_HEIGHT, IMG_WIDTH),color_mode='grayscale' ,batch_size=batch_size,seed=123,shuffle=True)
```

```python
val_data_gen=train_image_generator.flow_from_dataframe(ValDF1,x_col='filename',y_col='Target',target_size=(IMG_HEIGHT, IMG_WIDTH),color_mode='grayscale',batch_size=batch_size,seed=123,shuffle=True)
```

```python
test_data_gen=train_image_generator.flow_from_dataframe(TestDF1,x_col='filename',y_col='Target',color_mode='grayscale',batch_size=batch_size,target_size=(IMG_HEIGHT, IMG_WIDTH),seed=123,shuffle=True)
```
